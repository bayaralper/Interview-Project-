```python
import pandas as pd
```

Veri okuma

```python
walkers_df=pd.read_csv('Data Task//walkers.csv',index_col=0)
walks_df=pd.read_csv('Data Task//walks.csv',index_col=0)
```

```python
walkers_df
```

```python
walks_df
```

totalWalks verisinin eklenmesi #Bu esnada walkeridler de tutulmuştur ve diğer veriler buna göre eklenmiştir.

```python
totalWalks=walks_df.groupby('walkerid')['walkerid'].count()
```

```python
output_df=totalWalks.reset_index(name='totalWalks')
```

firstWalks verisinin eklenmesi

```python
firstWalks=walks_df.sort_values('checkintime',ascending=True).groupby('walkerid')['checkintime'].first().reset_index(name='firstWalks')
```

```python
output_df['firstWalks']=pd.to_datetime(firstWalks['firstWalks'])
```

```python
output_df
```

lastWalks verisinin eklenmesi

```python
lastWalks=walks_df.sort_values('checkintime',ascending=False).groupby('walkerid')['checkintime'].first().reset_index(name='lastWalks')
```

```python
output_df['lastWalks']=pd.to_datetime(lastWalks['lastWalks'])
```

signuptime verisinin eklenmesi

```python
walker_signtime=walkers_df[['id','signuptime']]
```

Walker datasetinde 730 adet kayıt bulunmasına rağmen walks datasetinde 1129 adet yürüyüş gerçekleştiren kişi vardır.Aynı zamanda bu 730 kişiden 355 adedinin yürüyüş bilgisi bulunmaktadır.Bu yüzden aşağıda birleştirme işlemi bu 730 kişi üzerinden gerçekleştirilmiş ve diğer hesaplamalarda bu 355 adet kişinin bilgileri kullanılmıştır.

```python
output_df=pd.merge(output_df,walker_signtime,how='right',left_on='walkerid',right_on='id')
```

```python
output_df
```

Bazı kullanıcıların walkerid'si bulunmadığı fakat id'si bulunduğu için ve outputta id istenmediği için walkerid kısmına NaN atmak veriyi belirsizleştireceği için çözüm üretmek amacıyla id ile eşitleme işlemi yaptım.(walks ve walker datasetinde id eşitliği göz önünde bulundurularak)

```python
output_df['walkerid']=output_df['id']
output_df=output_df.drop('id',axis=1)
```

```python
import numpy as np
```

```python
output_df['didDemandorWalk']=np.where(output_df['totalWalks'].isna()==False,'Yes','No')
```

```python
output_df['signuptime']=pd.to_datetime(output_df['signuptime'])
```

```python
output_df['lifetime']=np.where(output_df['lastWalks'].isna()==False,output_df['lastWalks']-output_df['signuptime'],np.nan)
```

```python
output_df['lifetime']=output_df['lifetime'].dt.days
```

```python
output_df
```

```python
walkercount=output_df[output_df['didDemandorWalk']=='Yes']['didDemandorWalk'].count()#355 kişi lifetime sahibi bu yüzden hesabı buna göre yapacağız.
```

```python
avglifetime=output_df['lifetime'].sum()/walkercount
```

```python
output_df['averageLifetime']=np.where(output_df['lifetime']>avglifetime,'More','Less')
```

walkercount walkerdatasetindeki 730 kişi ile sınırlandırılmıştır ve buradaki 355 kişi walks kayıtlarında aktiftir.

```python
avgwalkcount=output_df['totalWalks'].sum()/walkercount #Ortalama yürüyüş sayısı bulunur
```

```python
output_df['averageWalk']=np.where(output_df['totalWalks']>avgwalkcount,'More','Less') #Ortalama yürüyüş sayısına göre atama yapılır
```

```python
output_df[output_df['averageWalk']=='More']
```

```python
output_df
```

firstWalkAfterSignup sütununun bulunması

```python
output_df['firstWalkAfterSignup']=np.where(output_df['firstWalks'].isna()!=True,output_df['firstWalks']-output_df['signuptime'],np.nan)
```

```python
output_df['firstWalkAfterSignup']=output_df['firstWalkAfterSignup'].dt.days #Gün formatına çevrilir.
```

```python
import datetime
```

firstWalkinWeek sütununun bulunması

```python
output_df['firstWalkinWeek']=np.where(output_df['firstWalks'].isna()==True,np.nan,
        np.where((output_df['signuptime']+datetime.timedelta(days=7))<output_df['firstWalks'],'No','Yes'
                )
        )
#datetime.timedelta(days=7) ile kayıt tarihine 7 gün ekleme işlemi yaparak ilk hafta yürüme işlemi yapılıp yapılmadığına bakılır.
```

```python
output_df
```

```python
current_time=datetime.datetime.now() #Şu anın tarihi alınır.
last_month=current_time-datetime.timedelta(days=30)#Şu anki tarihten 30 gün öncesine gidilerek 1 ay öncesinin tarihi bulunur.
```

```python
output_df
```

```python
walks_df['checkintime']=pd.to_datetime(walks_df['checkintime'])
a=walks_df[walks_df['checkintime']>last_month].groupby('walkerid').count()
#checkintime 1 ay önceki tarihten daha yakın bir tarihse son 1 ayda yürüyüş yapıldığı anlamına gelir ve bunun sayısı bulunur.

a=a['duration']#Rastgele bir column alıyorum çünkü a count döndürdü
```

```python
#Eğer hiç yürüyüş yapılmamışsa 0 , yapılmışsa değer ataması yapılır.
if len(a)==0:
    output_df['lastMonthWalks']=0
else:
    output_df=output_df.merge(a,how='left')
    output_df['lastMonthWalks']=output_df['duration']
```

```python
output_df
```

activityTag sütunu bulunur

```python
output_df['activityTag']=np.where(output_df['lastWalks'].isna()==True,np.nan,
        np.where((current_time-datetime.timedelta(days=30))>output_df['lastWalks'],'No','Yes'
                )
        )
#Son bir ayda yürüyüş yapılıp yapılmadığı kontrol edilir.
```

differentDistrictCount sütunu bulunur

```python
district_counter=walks_df.groupby(['walkerid'])['district'].nunique().reset_index() #Unique districtler walkerid'ye göre bulunur.
```

```python
output_df=output_df.merge(district_counter,how='left')#Birleştirme işlemi yapılır.
```

```python
output_df['differentDistrictCount']=output_df['district']#İsim değişimi yapılır.
```

activeLifeTime sütunu bulunur.

```python
output_df['activeLifeTime']=output_df['lastWalks']-output_df['firstWalks']
```

```python
output_df
```

totalActiveDayCount sütunu bulunur.

```python
walks_df['day']=(pd.to_datetime(walks_df['endtime'])-walks_df['checkintime'])
```

```python
walks_df['day']=walks_df['day'].dt.days #Günler alınır.
```

```python
day_df=walks_df.groupby('walkerid')['day'].sum().reset_index()#Gün sayısı toplanır.
```

```python
output_df=output_df.merge(day_df,how='left')
```

```python
for i in output_df['day']:  #Bazı verilerde gün sayısı - olduğu için böyle bir değişikliğe gittim.Aynı günü 0 saydığı için 1 ekledim.
    if i<0:
        i=i*-1
    else:
        i+=1
output_df['totalActiveDayCount']=output_df['day'] #İsim değişimi
```

walkFrequency sütununun bulunması

```python
output_df['walkFrequency']=np.where(output_df['totalActiveDayCount'].isna()!=True,output_df['activeLifeTime']/output_df['totalActiveDayCount'],np.nan)
```

```python
output_df
```

dailyWalkAverage sütununun bulunması

```python
output_df['dailyWalkAverage']=output_df['totalWalks']/output_df['totalActiveDayCount']
```
